```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.describe()   # Mean, std, min, max, etc.
df.info()       # Data types and nulls
```

```python
df.hist(figsize=(10,8))
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title("Fare - Boxplot")
plt.show()

sns.boxplot(x=df['Age'])
plt.title("Age - Boxplot")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']])
plt.show()
```

```python
# Sirf numeric columns ke liye correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#  Step 6: Insights from visuals + data
```

```python
# 1. Pairplot  Check relationships between variables
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']])
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

```python
# 2. GroupBy for Pattern Checking
```

```python
# Pclass vs Survival:
df.groupby('Pclass')['Survived'].mean()
```

```python
# Sex vs Survival:
df.groupby('Sex')['Survived'].mean()
```

```python
# Embarked vs Survival:
df.groupby('Embarked')['Survived'].mean()
```

```python
# 3. Barplot - Male vs Female Survival
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
# 4. Age Distribution by Survival
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```
